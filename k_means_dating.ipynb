```python
#Packages
import pandas as pd
```

```python
df = pd.read_csv('speeddating.csv')
df.head(100)
```

```python
df.shape
print("Numbers of rows: ", df.shape[0])
print("Numbers of columns: ", df.shape[1])
```

```python
missing_values = df.isnull().sum()
percent_missing = (missing_values / len(df)) * 100
missing_data = pd.concat([missing_values, percent_missing], axis=1, keys=['Total', 'Percent'])
missing_data = missing_data.sort_values('Percent', ascending=False)
print(missing_data[:30])
```

```python
df1=df.drop(['expected_num_interested_in_me','expected_num_matches','shared_interests_o','shared_interests_partner','ambitous_o','ambition_partner','has_null'], axis=1)

print("After delete columns >5% missing values\n")
print("Numbers of rows: ", df1.shape[0])
print("Numbers of columns: ", df1.shape[1])

# Eliminamos todas las filas con valores nulos
df2 = df1.dropna(how='any')

print("\nAfter delete all empty rows\n")
print("Numbers of rows: ", df2.shape[0])
print("Numbers of columns: ", df2.shape[1])


print("\nPorcentage reduction data\n")
print("Reduction of rows: ", 100*(1-(df2.shape[0]/df.shape[0])))
print("Reduction of columns: ", 100*(1-(df2.shape[1]/df.shape[1])))
```

```python
suma_categorias = df2["match"].value_counts()
print(suma_categorias)
```

```python
# Convertir las variables categóricas en variables numéricas usando dummies
df2 = pd.get_dummies(df2)
df2.head(200)
```

```python
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')

# Separar los datos en características (X) y etiquetas (y)
df2_X= df2.drop("match_b'0'", axis=1)
df2_X= df2.drop("match_b'1'", axis=1)

X1 = df2_X
X2 = df2_X[["age"]]
y = df2[["match_b'0'"]]

from imblearn.over_sampling import RandomOverSampler

# Aplicar el sobremuestreo aleatorio a la clase minoritaria
ros = RandomOverSampler()
X_resampled, y_resampled = ros.fit_resample(X1, y)

ros = RandomOverSampler()
X2_resampled, y2_resampled = ros.fit_resample(X2, y)

print(len(X2))
print(len(X2_resampled))
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, accuracy_score, recall_score, precision_score, f1_score

X1 = df2_X

from imblearn.over_sampling import RandomOverSampler

# Aplicar el sobremuestreo aleatorio a la clase minoritaria
ros = RandomOverSampler()
X_resampled, y_resampled = ros.fit_resample(X1, y)

X1= X_resampled
y= y_resampled

# Normalizar las características en X1
scaler = StandardScaler()
X1_norm = scaler.fit_transform(X1)

# Realizar la clusterización con k-means
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X1_norm)
y_kmeans = kmeans.labels_


# Evaluar la calidad de la clusterización
silhouette = silhouette_score(X1_norm, y_kmeans)
print("Puntuación de Silhouette: {:.2f}".format(silhouette))

# Comparar las etiquetas obtenidas con las etiquetas originales
accuracy = accuracy_score(y, y_kmeans)
recall = recall_score(y, y_kmeans)
precision = precision_score(y, y_kmeans)
f1 = f1_score(y, y_kmeans)
print("Precisión: {:.2f}".format(accuracy))
print("Exhaustividad: {:.2f}".format(recall))
print("Precisión: {:.2f}".format(precision))
print("Puntuación F1: {:.2f}".format(f1))
```

```python
import matplotlib.pyplot as plt

# Realizar la clusterización con diferentes valores de k
inertias = []
for k in range(1, 25):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X1_norm)
    inertias.append(kmeans.inertia_)

# Graficar la suma de las distancias cuadradas en función de k
plt.plot(range(1, 25), inertias, 'bx-')
plt.xlabel('Número de clústeres (k)')
plt.ylabel('Inertia')
plt.title('Método del codo')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Reducir la dimensionalidad de los datos a 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X1_norm)

# Graficar los datos con los labels
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans, cmap='viridis')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('Clusterización con k-means')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# Reducir la dimensionalidad de los datos a 2D
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X1_norm)

# Graficar los datos con los labels
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y_kmeans, cmap='viridis')
plt.xlabel('Componente 1')
plt.ylabel('Componente 2')
plt.title('Clusterización con k-means')
plt.show()
```

```python
accuracy = accuracy_score(y, y_kmeans)
print("Precisión: {:.2f}".format(accuracy))
```

```python
from sklearn.model_selection import cross_val_score

# Definir el modelo de k-means
kmeans = KMeans(n_clusters=2, random_state=42)

# Calcular el score de validación cruzada para el modelo de k-means
scores = cross_val_score(kmeans, y, y_kmeans, cv=5, scoring='accuracy')

# Imprimir los resultados de validación cruzada
print("Scores de Validación Cruzada: {}".format(scores))
print("Precisión promedio: {:.2f}".format(scores.mean()))
```

```python
import matplotlib.pyplot as plt

# create a list of colors corresponding to the gender
colors = ['b' if g==0 else 'r' for g in df['gender_encoded']]

# plot the data points with different colors for males and females
plt.scatter(df['age'], df['gender_encoded'], c=colors)

# add axis labels
plt.xlabel('Age')
plt.ylabel('Gender (0=female, 1=male)')

# show the plot
plt.show()

```

```python
# group the data by age and gender, and count the number of males and females in each group
age_gender_counts = df.groupby(['age', 'gender'])['gender'].count()

# calculate the proportion of males and females in each age group
age_gender_proportions = age_gender_counts.groupby('age').apply(lambda x: x / float(x.sum()))

# create a bar chart of the proportion of males and females in each age group
ax = age_gender_proportions.unstack().plot(kind='bar', stacked=True)

# add axis labels
ax.set_xlabel('Age')
ax.set_ylabel('Proportion')

# move the legend to the top of the chart
ax.legend(['Female', 'Male'], loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=2)

# show the plot
plt.show()


```
